### 设置中文显示

```python
from matplotlib import font_manager
from matplotlib import pyplot as plt
x = range(2,26,2)
y = [15,13,14.5,17,20,25,26,26,24,22,18,15]
chfont = matplotlib.font_manager.FontProperties(fname="C:/Windows/Fonts/msyh.ttc")
plt.title("各时段温度变化",fontproperties=chfont)
plt.xlabel("时间",fontproperties=chfont)
plt.ylabel("温度",fontproperties=chfont)
plt.plot(x,y)
plt.show()
```

```python
from matplotlib import font_manager
from matplotlib import pyplot as plt
import random

myfont = font_manager.FontProperties(fname="C:/Windows/Fonts/msyh.ttc")
x = range(0,120)
y = [random.randint(20,35) for i in range(120)]
plt.figure(figsize=(20,8),dpi=80)
# 绘图
plt.plot(x,y)
# 调整x轴的刻度标签
xlabels = ["10点{}分".format(i) for i in range(60)]
xlabels += ["11点{}分".format(i) for i in range(60)]
# 取步长，数字和字符串一一对应，数据长度一致
plt.xticks(list(x)[::3],xlabels[::3],rotation=45,fontproperties=myfont)
# 添加描述信息
plt.xlabel("时间",fontproperties=myfont)
plt.ylabel("温度",fontproperties=myfont)
plt.title("10点到12点每分钟气温变化情况",fontproperties=myfont,size=18)
plt.show()
```
