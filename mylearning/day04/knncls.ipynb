### K-近邻算法，距离为“欧氏距离”，优缺点如下：

优点：简单，易于理解，易于实现，无需估计参数，无需训练（因为所给出特征矩阵各特征列元素之间的欧氏距离是一定的）

缺点：懒惰算法，对测试样本分类是的计算量大，内存开销大；必须指定K值，K值选择不当则分类精度不能保证
